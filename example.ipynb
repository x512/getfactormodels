## Examples

### 1. Using `get_factors`

`get_factors` params: model, frequency, start_date, end_date, output_file.

```python
import getfactormodels

df = getfactormodels.get_factors(model="carhart", frequency='m')
print(df)
```

```python
import getfactormodels as gfm

data = gfm.get_factors(model="3", frequency='d', 
                       start_date="1990-01-01", end_date="2010-12-31")
print(data)
```

```python
data = gfm.get_factors(model="3", 
                       frequency='m', 
                       start_date="1965-01-01")
print(data)
```

### 1. Classes

```python
import getfactormodels as gfm

q = gfm.QFactors(frequency='d', start_date='2020-01-01', end_date='2021-12-31')
q_df = q.download()

print(q_df)
```

```python
from getfactormodels import FamaFrenchFactors

model = FamaFrenchFactors(model=3, frequency='d')
data = model.download()
```

```python
print(data)
```

```python
from getfactormodels import QFactors
data = QFactors(frequency='d')

df = data.download()

print(df)
```

```python
from getfactormodels import MispricingFactors

data = MispricingFactors(start_date="2015-01-01", end_date="2019-12-31") # frequency = m by default

df = data.download()
```

```python
print(df)
```
