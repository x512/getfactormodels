```python
from getfactormodels import ff_factors, carhart_factors

# 3, 5, 6 Monthly
df3 = ff_factors(model="3", frequency="M", start_date="1920-01-01")
df4 = carhart_factors(freq="M", start_date="1920-01-01")  # Note: if '4' is passed to ff_factors, it returns the carhart_factors
# [TODO]: 'freq' should be changed to 'frequency' to be consistent
df5 = ff_factors(model="5", frequency="M", start_date="1920-01-01")
df6 = ff_factors(model="6", frequency="M", start_date="1920-01-01")

print(df3.head())
print(df4.head())
print(df5.head())
print(df6.head())
```

```python
from getfactormodels import icr_factors, hml_devil_factors

icr = icr_factors(frequency="M", start_date="1970-01-01")
print(icr.head())

icr_daily = icr_factors(frequency="D", start_date="1999-05-03")
print(icr_daily.head())
```

* Note: ICR daily factors start on 1999-05-03. NaN in Risk Factor on first day.
