## CLI

Example notebook showing how to use `getfactormodels` from the command-line.

A quick and minimal how-to, showing some features.

#### Get a model

- Use the `--model`/`-m` parameter with a model name, and optionally specify a frequency with `--frequency/-f`. By default, monthly `m` data is returned.

```python
%%bash
getfactormodels --model ff3 --frequency m
```

- Specify a date range using `-s`/`--start` and `-e`/`--end`. Date input accepts `YYYY`, `YYYY-MM`, or `YYYY-MM-DD`. All dates are inclusive.

```python
%%bash
getfactormodels --model ff5 --frequency d --start 2000 --end 2010-12-31
```

- You can save to file with the `-o`/`--output` param. Accepts a filename or path with extension. Supports .csv, .txt, .parquet, .ipc, .md.
- Using `--quiet`/`-q` supresses all output to console.

```python
%%bash
getfactormodels -m q -f m -s 1980-01 -e 1980-12 --output file.md --quiet

# NOTE: e.g., -m q -s 1960-01 -e 1960-12
#   q doesn't have data here, so it errors, handle this error.
```

- Preview of the markdown file:

```python
# This prints the markdown file to the cell below
from IPython.display import Markdown, display
with open('file.md', 'r', encoding='utf-8') as f:
    content = f.read()
display(Markdown(content))
```

- Get multiple models of the same frequency

```python
%%bash
getfactormodels -m ff3 liq -f m -s 2000 -e 2010
```

- Extract or drop factor(s) from a model with `--drop`/`-d` and `--extract`/`-x`

```python
%%bash
getfactormodels -m carhart -f d -x MOM
```

```python
%%bash
getfactormodels -m ff6 -d y -d RF HML
```

#### Get a portfolio

- Use the `--portfolio`/`-p` param to download portfolio return data from the Kenneth French Data Library. Supports all industry portfolios and US sorts. Note: `--portfolio` or `--model` are required.
- To download industry return data, you can use `--industry`/`-I` param with any of the available industry portfolios available, `5`, `12`, `20`, `30`, `32` ... 

```python
%%bash
getfactormodels --industry 5
# Equiv to:
#getfactormodels --portfolio industry 5
```

```python
%%bash
# 10 portfolios sorted on Book-to-Market
getfactormodels --portfolio bm -n 10
#getfactormodels -p bm --sort decile
```

```python
%%bash
# Bivariate: 6 portfolios sorted on size and OP
# --sort/-n accepts the sort (2x3) or portfolios (6), and --weights/-W accepts eithers value `vw` (default) or equal `ew`
getfactormodels --portfolio size op -n 6
getfactormodels -p op,size -n 5x5
```

- Attach a portfolio of returns to a model:

```python
%%bash
getfactormodels --model ff3 --frequency m --portfolio op -n 5 --start 1990 --end 1995-03
```
